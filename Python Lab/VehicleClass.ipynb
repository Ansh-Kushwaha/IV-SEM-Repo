##16.Write a Python program to create a VEHICLE class and child classes that inherit its instance attributes.

```python
class Vehicle:
    def __init__(self, brand_name, vehicle_type, size, fuel_type):
        self.brand_name = brand_name
        self.vehicle_type = vehicle_type
        self.size = size
        self.fuel_type = fuel_type
    
    def display(self):
        print("Brand Name:", self.brand_name)
        print("Type:", self.vehicle_type)
        print("Size:", self.size)
        print("Fuel Type:", self.fuel_type)


class Car(Vehicle):
    def __init__(self, brand_name, size, fuel_type):
        super().__init__(brand_name, "Four Wheeler", size, fuel_type)


class Motorcycle(Vehicle):
    def __init__(self, brand_name, size, fuel_type):
        super().__init__(brand_name, "Two Wheeler", size, fuel_type)


car1 = Car("Toyota", "Large", "CNG")
car1.display()

print()

motorcycle1 = Motorcycle("Suzuki", "Small", "Petrol")
motorcycle1.display()
```
