#Overview of casting

```python
a = input('Enter any number: ')
b = input('Enter another number: ')
sum = int(a) + int(b)
print('Sum of a and b:', sum)
print('Data type of sum:', type(sum))
```

```python
con = str(a) + str(b)
print('Concatenation of a and b:', con)
print('Data type of con:', type(con))
```
