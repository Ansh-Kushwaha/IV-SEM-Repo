```python
i = 1
while i < 4 :
  print('Outer Loop Iteration: ', i)
  i += 1
```

```python
i = 1
while i <= 10:
  if i == 5:
    i += 1
    continue
  elif i == 8:
    break
  print(i)
  i += 1
else: 
  print('i is greater than 7')
```
