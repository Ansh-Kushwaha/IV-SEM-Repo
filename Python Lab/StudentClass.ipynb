##15.Write a Python program to create a STUDENT class with instance attributes like enrollment numver and branch name etc.

```python
class Student:
    def __init__(self, enrollment_number, branch_name, name, age):
        self.enrollment_number = enrollment_number
        self.branch_name = branch_name
        self.name = name
        self.age = age
    
    def display_student_details(self):
        print("Enrollment Number:", self.enrollment_number)
        print("Branch Name:", self.branch_name)
        print("Name:", self.name)
        print("Age:", self.age)

s1 = Student('0103CS211030', 'CSE', 'Ansh Kushwaha', 19)
s1.display_student_details()
```
