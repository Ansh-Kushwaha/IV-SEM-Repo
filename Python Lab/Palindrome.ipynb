##10.Write a program to print the integer is palindrome is not.

```python
def palindrome(n):
  lst = []
  while n != 0:
    d = n % 10
    lst.append(d)
    n = n // 10
  l = len(lst)
  m = l // 2
  
  for i in range(0, m):
    if lst[i] != lst[l - i - 1]:
      return False
  return True

n = int(input('Enter a number: '))
if palindrome(n):
  print(n, 'is palindrome')
else:
  print(n, 'is not palindrome')
```
