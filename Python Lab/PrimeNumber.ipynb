##12.Write a program find the given number is prime or not.

```python
from math import sqrt

def isPrime(n):
  if n <= 1:
    return False
  else:
    for i in range(2, int(sqrt(n)) + 1):
      if n % i == 0:
        return False
    return True
  
n = int(input('Enter a number: '))

if isPrime(n):
  print(n, 'is a prime number.')
else:
  print(n, 'is not a prime number')
```
